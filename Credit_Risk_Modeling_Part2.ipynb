```python
import pandas as pd

file_path = '/content/drive/MyDrive/Personal_Project_1/final_predictions.csv'
df = pd.read_csv(file_path)
df.head()
```

```python
df.columns
```

```python
from sklearn.metrics import mean_absolute_error, accuracy_score

# Store results
results = {}

# Loop through models
for model in ['xgb', 'rf', 'lr']:
    # Probabilities and predictions
    prob_col = f'{model}_prob'
    pred_col = f'{model}_pred'

    mae = mean_absolute_error(df['default_flag'], df[prob_col])
    accuracy = accuracy_score(df['default_flag'], df[pred_col])

    results[model.upper()] = {
        'MAE': round(mae, 4),
        'Accuracy': round(accuracy, 4)
    }

# Display results
import pandas as pd
results_df = pd.DataFrame(results).T
results_df
```
